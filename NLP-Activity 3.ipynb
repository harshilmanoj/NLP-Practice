```python
import nltk
from nltk.corpus import words
nltk.download('words')
string = "thisisinsane"
token =[]
lowerCorpus = [x.lower() for x in words.words()]
i = 0
while i <len(string):
  maxword = ""
  for j in range(i,len(string)):
    tempword = string[i:j+1]
    if tempword in lowerCorpus and len(tempword) > len(maxword):
      maxword = tempword
  i = i+len(maxword)
  token.append(maxword)
print(token)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer

textporter = 'VIT was established with the aim of providing quality higher education on par with international standards. It persistently seeks and adopts innovative methods to improve the quality of higher education on a consistent basis. The campus has a cosmopolitan atmosphere with students from all corners of the globe. Experienced and learned teachers are strongly encouraged to nurture the students. The global standards set at VIT in the field of teaching and research spur us on in our relentless pursuit of excellence. In fact, it has become a way of life for us. The highly motivated youngsters on the campus are a constant source of pride. Our Memoranda of Understanding with various international universities are our major strength. They provide for an exchange of students and faculty and encourage joint research projects for the mutual benefit of these universities. Many of our students, who pursue their research projects in foreign universities, bring high quality to their work and esteem to India and have done us proud. With steady steps, we continue our march forward. We look forward to meeting you here at VIT'

tokensforporter = word_tokenize(textporter)
tokensforporter = list(set(tokensforporter))
print("List of words in normal text : ")
print(tokensforporter,"\n")

porterstemmer = PorterStemmer()
stemsporter = [porterstemmer.stem(w) for w in tokensforporter]
stemsporter = list(set(stemsporter))
print("List of words in stemmed text : ")
print(stemsporter,"\n")

print("No. of words in normal text : ",len(tokensforporter))
print("No. of words in stemmed text : ",len(stemsporter),"\n")
print("Ratio of the total number of words present in the original text to that in the stemmed text using PorterStemmer : ",len(stemsporter)/len(tokensforporter),"\n")
```
