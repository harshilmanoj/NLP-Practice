**POS tagging for "shakespeare-caesar.txt‚Äù from Gutenberg Corpus**

**Method : Averaged_perceptron_tagger**

```python
import nltk
nltk.download('punkt')
from nltk.corpus import gutenberg
nltk.download('averaged_perceptron_tagger')
nltk.download('gutenberg')
gwords = set(gutenberg.words("shakespeare-caesar.txt"))
nltk.pos_tag(gwords)
```

**Using the Tagged Corpus to find the conditional probability distribution of words.**

**Probability of an adverb being ‚Äònot‚Äô and a determinant being ‚Äòthe‚Äô:**

```python
gutenberg_tagged_sents = [nltk.pos_tag(gwords)]
gutenberg_word_tags=[]
for gutenberg_sent in gutenberg_tagged_sents:
    gutenberg_word_tags.append(('START','START'))
    for words,tag in gutenberg_sent:
      gutenberg_word_tags.extend([(tag[:2],words)])
    gutenberg_word_tags.append(('END','END'))


cfd1_tag_words_guten=nltk.ConditionalFreqDist(gutenberg_word_tags)
cpd_tag_words_guten=nltk.ConditionalProbDist(cfd1_tag_words_guten,nltk.MLEProbDist)


print("The probability of an adverb (RB) being 'not' is", cpd_tag_words_guten["RB"].prob("not"))
print("The probability of a determinant (DT) being 'the' is", cpd_tag_words_guten["DT"].prob("the"))
```

**Probability of DT occurring after NN and VB occurring after NN:**

```python
guten_tags=[]
for tag, words in gutenberg_word_tags:
    guten_tags.append(tag)

cfd_tags_gutenberg=nltk.ConditionalFreqDist(nltk.bigrams(guten_tags))
cpd_tag_gutenberg=nltk.ConditionalProbDist(cfd_tags_gutenberg,nltk.MLEProbDist)

print('The probability of DT occuring after NN is : ', cpd_tag_gutenberg["NN"].prob("DT"))
print('The probability of VB occuring after NN is : ', cpd_tag_gutenberg["NN"].prob("VB"))
```

**Probability of an adjective being ‚Äòsmart‚Äô and a verb being ‚Äòtry‚Äô:**

```python
from nltk.corpus import brown
brown.tagged_sents()
brown_word_tags=[]
for brown_sent in brown.tagged_sents():
    brown_word_tags.append(('START','START'))
    for words,tag in brown_sent:
      brown_word_tags.extend([(tag[:2],words)])
    brown_word_tags.append(('END','END'))

cfd_tag_words=nltk.ConditionalFreqDist(brown_word_tags)
cpd_tag_words=nltk.ConditionalProbDist(cfd_tag_words,nltk.MLEProbDist)
print("The probability of an adjective (JJ) being 'smart' is", cpd_tag_words["JJ"].prob("smart"))
print("The probability of a verb (VB) being 'try' is", cpd_tag_words["VB"].prob("try"))
```

**Probability of DT occurring after NN and VB occurring after NN:**

```python
brown_tags=[]
for tag, words in brown_word_tags:
    brown_tags.append(tag)
cfd_tags=nltk.ConditionalFreqDist(nltk.bigrams(brown_tags))
cpd_tags=nltk.ConditionalProbDist(cfd_tags,nltk.MLEProbDist)
print('The probability of DT occuring after NN is : ', cpd_tags["NN"].prob("DT"))
print('The probability of VB occuring after NN is : ', cpd_tags["NN"].prob("VB"))
```
