```python
sent_tokenize(text)
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize(text)
sent_tokenize('Hi,how are you?"I am good,you?"Great!')
```

```python
from nltk.tokenize import word_tokenize
word_tokenize(text)
word_tokenize("Hi,how are you?I'am good,you?Great!")
```

```python
grammar1 = nltk.CFG.fromstring("""
S -> NP VP
VP -> V NP | V NP PP
PP -> P NP
V -> "saw" | "ate"
NP -> "Sue" | "John" | Det N | Det N PP
Det -> "a" | "the" | "an"
N -> "mouse" | "cat" | "dog" | "park" | "telescope"
P -> "in" |"on" | "by"
""")
```

```python
from nltk.tree import Tree
from IPython.display import display

sent = "Sue saw a mouse".split()
rd_parser = nltk.RecursiveDescentParser(grammar1)
for tree in rd_parser.parse(sent):
    print(tree)
display(tree)
```
