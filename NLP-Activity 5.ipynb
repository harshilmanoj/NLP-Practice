**Chunking : 
Applying noun phrase and verb phrase chunking to a few sentences.**

```python
from nltk import pos_tag
from nltk import RegexpParser
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

**GRAMMAR FOR NOUN PHASE**

```python
grammarnp="NP:{<DT>?<JJ>*<NN>}"
chunknp = RegexpParser(grammarnp)
```

**SENTENCE 1**


```python
sentencenp1 ="My name is Harshil."
tokennp1 = word_tokenize(sentencenp1)
tagnp1 = pos_tag(tokennp1)
print(tagnp1)
outputnp1 = chunknp.parse(tagnp1)
print()
print(outputnp1)
print()
outputnp1.draw
```

**SENTENCE 2**

```python
sentencenp2 ="I am a Third year Computer Science Engineering Student."
tokennp2 = word_tokenize(sentencenp2)
tagnp2 = pos_tag(tokennp2)
print(tagnp2)
outputnp2 = chunknp.parse(tagnp2)
print()
print(outputnp2)
print()
outputnp2.draw
```

**SENTENCE 3**

```python
sentencenp3 ="I am very motivated for my final year."
tokennp3 = word_tokenize(sentencenp3)
tagnp3 = pos_tag(tokennp3)
print(tagnp3)
outputnp3 = chunknp.parse(tagnp3)
print()
print(outputnp3)
print()
outputnp3.draw
```

**GRAMMAR FOR VERB PRHASE**

```python
grammarvp="VP: {<MD>?<VB.*><NP|PP>} "
chunkvp = RegexpParser(grammarvp)
```

**Lesk Algorithm : Apply Lesk algorithm to sentences that contain a word with multiple word senses.**

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
sen1 = lesk(word_tokenize('I left my phone on the left side of the room.'),'left')
print(sen1,sen1.definition())
```

```python
sen2 = lesk(word_tokenize('She will park the car so we can walk in the park'),'park')
print(sen2,sen2.definition())
```

**Spacy : Explore spacy package for nlp**

```python
import nltk
import spacy 
nlp = spacy.load("en_core_web_sm")


text = ('My name is Harshil'' I am using Spacy for nlp activity.')
doc = nlp(text)
print ([token.text for token in doc])
```

```python
about_text = ('My name is Harshil . I am a Computer Science Engineering Student.' 'I am using Spacy for NLP activity')
about_doc = nlp(about_text)
sentences = list(about_doc.sents)
len(sentences)
for sentence in sentences:
    print (sentence)
```
