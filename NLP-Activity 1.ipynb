```python
import nltk
import matplotlib.pyplot as plt
import numpy as np
nltk.download('punkt')
nltk.download('gutenberg')
from nltk.corpus import gutenberg
gutenberg.fileids()
```

```python
emma = gutenberg.words('austen-emma.txt')
emma
```

```python
print("File_Id",'->',"No. of Characters",'->',"No. of Words",'->',"No. of Sentences",'->',"No. of Vocabulary",'->',"Avgerage word length",'->',"Average Words per Sentence")
for fileid in gutenberg.fileids():
    numofchars = len(gutenberg.raw(fileid))
    numofwords = len(gutenberg.words(fileid))
    avgwordlen = round(numofchars/numofwords)
    numofsents = len(gutenberg.sents(fileid))
    numofvocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    avgwordlen = round(numofchars/numofwords)
    avgwordspersent = round(numofwords/numofsents)
    print(fileid,'->',numofchars,'->',numofwords,'->',numofsents,'->',numofvocab,'->',avgwordlen,'->',avgwordspersent)
    
```

```python
textguten = gutenberg.words("shakespeare-macbeth.txt")
fgutendist = nltk.FreqDist(w.lower() for w in textguten)
fgutendist
```

```python
modelgutenlist = nltk.FreqDist(w.lower() for w in textguten[:50])
modelgutenlist.plot()
```
